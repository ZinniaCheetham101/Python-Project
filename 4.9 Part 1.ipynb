```python
# Create a subset with the missing values
df_nan = df_customers[df_customers['First Name'].isnull() == True]
```

```python
#Create a clean dataframe without NaN values
df_customers_clean= df_customers[df_customers['First Name'].isnull() == False]
```

```python
#Compare the shape of the new dataframe to the original
df_customers_clean.shape
```

```python
# Look for full duplictates within your data frame
df_dups = df_customers_clean[df_customers_clean.duplicated()]
```

```python
df_dups
```

There are no duplicates in this data set and NaN values have been omitted. 

```python
#Check for mixed-type data in your df_customers_clean dataframe
for col in df_customers_clean.columns.tolist():
  weird = (df_customers_clean[[col]].applymap(type) != df_customers_clean[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers_clean[weird]) > 0:
    print (col)
```

No columns were printed so there are no mixed-type data detected
