```python
#Import Pickle File
#Plot a histogram
orders_products_all=pd.read_pickle(os.path.join(path, 'Data','Prepared Data', 'orders_products_all.pkl'))
```

```python
orders_products_all.columns
```

```python
#Plot a histogram
orders_products_all['order_hour_of_day'].plot.hist(bins = 24)
```

A histogram visualized frequency. In this chart it is showing the frequency of different hours of the day for orders. This histogram shows that sales are low from Midnight-5 AM then increase steadily from 6-9AM, followed by the peak frequency from 10-3PM then there is a steady decrease from 4-11pm. Mid-day is therefore the busiest time according to this chart

```python
#Create a bar chart to show the distribution of orders for customer loyalty categories
orders_products_all['loyalty_flag'].value_counts().plot.bar()
```

```python
line = sns.lineplot(data = orders_products_all, x = 'order_hour_of_day', y = 'prices')
```

The line chart shows that expenditure does fluctuate during the day with a drop around 9-10AM. Maybe that could indicate breakfast sales instead of groceries?

```python
#Make a line chart for age and number of dependents
line_Age = sns.lineplot(data = orders_products_all, x = 'n_dependants', y = 'Age')
```

This graph does not support a correlation (positive or negative) between age and the number of dependents. 

```python
#Make scatterplot for age and spending power 'income'
sns.scatterplot(x = 'income', y = 'Age',data = orders_products_all)
```

This scatterplot shows that income (which directly relates to spending power) is generally evenly spread across all age groups up to 200,000. After that income tends to be correlated with age with more individuals over 40 earning between 200,000-400,000 than those below 40. In addition, only individuals over 40 have an income of over 400,000.
