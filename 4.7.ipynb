```python
#Task 4.7 Create column for two busiest days of the week using for-loop

result_days_of_week = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value ==0:
    result_days_of_week.append("Busiest days")
  elif value ==1:
    result_days_of_week.append ("Busiest days")
  elif value == 4:
    result_days_of_week.append ("Least busy days")
  elif value ==3:
    result_days_of_week.append("Least busy days")
  else:
    result_days_of_week.append("Regularly busy")
```

```python
#Find busiest and slowest times for orders by hour in the dataset
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna=False)
```

```python
#Create column for busiest times of the day using for-loop
#Hours with over 2,500,000 orders were placed in "Most Orders". 
#Hours with below 100,000 orders were placed in "Fewest Orders"

result_hours_of_the_day = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value ==10:
    result_hours_of_the_day.append("Most Orders")
  elif value ==11:
    result_hours_of_the_day.append ("Most Orders")
  elif value ==14:
    result_hours_of_the_day.append ("Most Orders")
  elif value ==15:
    result_hours_of_the_day.append ("Most Orders")
  elif value ==13:
    result_hours_of_the_day.append ("Most Orders")
  elif value ==12:
    result_hours_of_the_day.append ("Most Orders")
  elif value ==16:
    result_hours_of_the_day.append ("Most Orders")  
  elif value == 5:
    result_hours_of_the_day.append ("Fewest Orders")
  elif value ==2:
    result_hours_of_the_day.append("Fewest Orders")
  elif value == 4:
    result_hours_of_the_day.append ("Fewest Orders")
  elif value ==3:
    result_hours_of_the_day.append("Fewest Orders") 
  else:
    result_hours_of_the_day.append("Average Orders")
```

```python
result_hours_of_the_day
```

```python
#Create new column in dataframe
df_ords_prods_merged['busiest hours of the day']=result_hours_of_the_day
```

```python
#Print the frequency of the new column
df_ords_prods_merged['busiest hours of the day'].value_counts(dropna=False)
```
