```python
#Importing Libraries
import pandas as pd
import numpy as np
import os
```

```python
path= r'C:\Users\Zinnia\Documents\Career Foundry Data Analytics Course\Achievement 4\A4 Project\Data'
```

```python
#Import Data
ords_prods_merge=pd.read_pickle(os.path.join(path, 'Prepared Data', 'orders_products_merged_june.pkl'))
```

```python
#Find the aggregated mean of the order_id column grouped by department_id
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

#Step 3 Result Analysis
#The mean is slightly different since this mean takes into account every recorded order instead of just utilizing those in the subset we selected. The numbers did not change greatly though. In the last set, the order mean was between 14-26, while in this set its 15-22.

```python
#Create a new max_order column grouped by user_id
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
#Use loc() to apply flag to different loyalty customer groups
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Review changes by checking value counts in new column
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge.columns
```

```python
#Use the loyalty flag column to check the statistics of prices for each loyalty category
ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean']})
```

```python
#Split the data into groups based on the User-id column
ords_prods_merge.groupby('user_id').agg({'prices': ['mean']})
```

```python
#Apply the transform function to generate data for each user and create a new column for the results of the aggregation
ords_prods_merge['users_average_order'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Check that new users_average_order column was created
ords_prods_merge.head()
```

```python
#use the new column to assign a spending flag
ords_prods_merge.loc[ords_prods_merge['users_average_order'] >=10, 'spending_flag'] = 'High Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['users_average_order'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
#Check data by using head() function on first 60 rows for new columns
ords_prods_merge[['users_average_order', 'spending_flag']].head(60)
```

```python
#Split the data into groups based on the days_since_prior_order column
ords_prods_merge.groupby('user_id').agg({'days_since_prior_order': ['mean']})
```

```python
#Apply the transform function to generate data for each user and create a new column for the results of the aggregation
ords_prods_merge['users_days_since_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.mean)
```

```python
#Check that new users_days_since_order column was created
ords_prods_merge.head()
```

```python
#use the new column to assign an order frequency flag
ords_prods_merge.loc[ords_prods_merge['users_days_since_order'] >20, 'Order_frequency_flag'] = 'Non-frequent Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['users_days_since_order'] >10) & (ords_prods_merge['users_days_since_order'] <=20),'Order_frequency_flag'] = 'Regular Customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['users_days_since_order'] <=10, 'Order_frequency_flag'] = 'Frequent Customer'
```

```python
#Check data by using head() function on first 60 rows for new columns
ords_prods_merge[['users_days_since_order', 'Order_frequency_flag']].head(60)
```

```python
#Export data in pickle format
ords_prods_merge.to_pickle(os.path.join(path,'Prepared Data', 'orders_products_merged_4.8.pkl'))
```
